```python
# click_log_df.groupby('c')
# click_log_df_merge.groupby('user_id')['words_count'].agg({'mean'}).sort_values('mean', ascending=False).reset_index()['mean'].plot()
def mean_diff_time_func(df, col):
    print(df, col)
    df = pd.DataFrame(df, columns={col})
    df['time_shift1'] = df[col].shift(1).fillna(0)
    df['diff_time'] = abs(df[col] - df['time_shift1'])
    return df['diff_time'].mean()

from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
click_log_df_merge['click_timestamp'] = mm.fit_transform(click_log_df_merge[['click_timestamp']])
click_log_df_merge['created_at_ts'] = mm.fit_transform(click_log_df_merge[['created_at_ts']])
mean_diff_click_time = click_log_df_merge.groupby('user_id')['click_timestamp', 'created_at_ts'].apply(lambda x: mean_diff_time_func(x, 'click_timestamp'))
```

```python
path = 'C:/ZhangLI/Codes/DataSet/新闻推荐比赛数据/'
train = path + 'articles.csv'
train_click = path + 'train_click_log.csv'
test_click = path + 'testA_click_log.csv'

item_info_df = pd.read_csv(train)
click_log_df = pd.read_csv(train_click)
# test_click_log_df = pd.read_csv(test_click)
```

```python
# 【召回评估】
# debug：get all click sample
all_click = pd.read_csv(train_click)  # 20w users, 111w rows
all_user_ids = all_click.user_id.unique()  # nunique->cnt unique->array
sample_user_ids = np.random.choice(all_user_ids, size=100)  # array->array
all_click = all_click[all_click.user_id.isin(sample_user_ids)]  # df->df
all_click = all_click.drop_duplicates(subset=['user_id', 'click_article_id', 'click_timestamp'])
# online:
pass
# 【数据处理】
max_min_scaler = lambda x: (x-np.min(x))/(np.max(x)-np.min(x))
all_click['click_timestamp'] = all_click[['click_timestamp']].apply(max_min_scaler)
```

```python
# 基于关联规则的协同过滤会用到
# 【user: article-time】
all_click = all_click.sort_values(['click_timestamp'])
def make_item_time_pair(df):
    return list(zip(df['click_article_id'], df['click_timestamp']))
user_item_time_df = all_click.groupby('user_id')['click_article_id', 'click_timestamp'].apply(lambda x: make_item_time_pair(x)).reset_index().rename(columns={0: 'item_time_list'})
user_item_time_dict = dict(zip(user_item_time_df['user_id'], user_item_time_df['item_time_list']))
# 【article: user-time】
def make_user_time_pair(df):
    return list(zip(df['user_id'], df['click_timestamp']))
item_user_time_df = all_click.groupby('click_article_id')['user_id', 'click_timestamp'].apply(lambda x: make_user_time_pair(x)).reset_index().rename(columns={0: 'user_time_list'})
item_user_time_dict = dict(zip(item_user_time_df['click_article_id'], item_user_time_df['user_time_list']))
# 【history and the last click】
click_last_df = all_click.sort_values(['user_id', 'click_timestamp']).groupby('user_id').tail(1)  # return last n rows of each group
def history_func(x):
    if len(x)==1:
        return x
    else:
        return x[:-1]
click_hist_df = all_click.groupby('user_id').apply(history_func).reset_index(drop=True)  # return history except last n rows
# 【article infomation】
item_info_df[['created_at_ts']] = item_info_df[['created_at_ts']].apply(max_min_scaler)
item_info_df = item_info_df.rename(columns={'article_id':'click_article_id'})
item_type_dict = dict(zip(item_info_df['click_article_id'], item_info_df['category_id']))
item_word_dict = dict(zip(item_info_df['click_article_id'], item_info_df['words_count']))
item_created_time_dict = dict(zip(item_info_df['click_article_id'], item_info_df['created_at_ts']))
# 【history infomation】
# [1. user-category]
user_hist_item_typs = all_click.groupby('user_id')['category_id'].agg(set).reset_index()
user_hist_item_typs_dict = dict(zip(user_hist_item_typs['user_id'], user_hist_item_typs['category_id']))
# [2. user-article]
user_hist_item_ids = all_click.groupby('user_id')['click_article_id'].agg(set).reset_index()
user_hist_item_ids_dict = dict(zip(user_hist_item_ids['user_id'], user_hist_item_ids['click_article_id']))
# [3. user-word_cnt]
user_hist_item_words = all_click.groupby('user_id')['words_count'].agg('mean').reset_index()
user_hist_item_words_dict = dict(zip(user_hist_item_words['user_id'], user_hist_item_words['words_count']))
# [4. user-last_time]
all_click_ = all_click.sort_values('click_timestamp')
user_last_item_created_time = all_click_.groupby('user_id')['created_at_ts'].apply(lambda x: x.iloc[-1]).reset_index()
max_min_scaler = lambda x : (x-np.min(x))/(np.max(x)-np.min(x))
user_last_item_created_time['created_at_ts'] = user_last_item_created_time[['created_at_ts']].apply(max_min_scaler)
user_last_item_created_time_dict = dict(zip(user_last_item_created_time['user_id'], user_last_item_created_time['created_at_ts']))
# 【top-click】
topk_click = all_click.groupby('click_article_id')['user_id'].count().sort_values().index[-10:]
```
